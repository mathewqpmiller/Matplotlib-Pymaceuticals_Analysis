## Pymaceuticals Analysis

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files:
# Define variables that hold the csv path location
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read:
# Define variables that read the csv path location variables into jupyter notebook
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
mouse_metadata.head()
```

```python
study_results.head()
```

```python
# Merge:
# Combine the data into a single dataset
# Define variable and use pandas merge function (pd.merge) on "Mouse ID" so that no data is lost
merged_mouse_dataset = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
# Display the data table for preview
merged_mouse_dataset.head()
```

```python
# Unique count:
# Check the number of mice.
# Use pandas function (.nunique) to count the number of distinct observations in the merged variable. (There are 249 ID's)
merged_mouse_dataset.nunique()
```

```python
# Print mouse count findings statment
print("There are 249 unique mouse ID's")
```

```python
# DataFrame:
# Getting the Mouse ID numbers that show up with duplicate Timepoints. 
# Create a DataFrame that holds just the Mouse ID and Timepoints and assign it a variable name
mouse_timepoint_duplicates = pd.DataFrame(merged_mouse_dataset, columns=['Mouse ID', 'Timepoint'])
# Print the DataFrame
mouse_timepoint_duplicates
```

```python
# Duplicates:
# Use pandas function find duplicate (.duplicated) with the duplicates DataFrame variable to identify repeats
duplicate_results = mouse_timepoint_duplicates[mouse_timepoint_duplicates.duplicated()]
# Create and print a header description for results
print("Duplicate Mouse ID/Timepoint rows, except first instances, are:")
# Print the findings
print(duplicate_results)
```

```python
# Locate:
# Optional: Display all the data for the duplicate mouse ID. 
# With the original merged dataset variable, use pandas locate (.loc) to find the desired column and associated row
merged_mouse_dataset.loc[merged_mouse_dataset['Mouse ID'] == 'g989']
```

```python
# Index / Delete:
# Create a clean DataFrame by dropping all data associated with the duplicate mouse by its ID.
# Index a row based on value 'g989' in the dataset and define it with a variable "delete_index_g989"
delete_index_g989 = merged_mouse_dataset[merged_mouse_dataset["Mouse ID"]== 'g989'].index
# Use pandas drop function (.drop), with the merged dataset variable, to delete the indexed variable from the dataset
merged_mouse_dataset = merged_mouse_dataset.drop(delete_index_g989)
merged_mouse_dataset.head()
```

```python
# Count:
# Checking the number of mice in the clean DataFrame.
merged_mouse_dataset.nunique()
```

```python
# Print a new mouse count findings statment after dataset was cleaned
print("There are now 248 unique mouse ID's")
```

## Summary Statistics

```python
merged_mouse_dataset['Drug Regimen'].nunique()
```

```python
# DataFrame:
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
drug_regimen_group = merged_mouse_dataset.set_index('Drug Regimen').groupby(['Drug Regimen'])
# Define variables and use pandas functions to find mean, median, variance, standard deviation, and SEM of the tumor volume.
mouse_tumor_volume_mean = drug_regimen_group['Tumor Volume (mm3)'].mean()
mouse_tumor_volume_median = drug_regimen_group['Tumor Volume (mm3)'].median()
mouse_tumor_volume_variance = drug_regimen_group['Tumor Volume (mm3)'].var()
mouse_tumor_volume_standard_deviation = drug_regimen_group['Tumor Volume (mm3)'].std()
mouse_tumor_volume_standard_error = drug_regimen_group['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary dataframe.
mouse_tumor_volume_summary_statistics = pd.DataFrame({"Mean":mouse_tumor_volume_mean,"Median":mouse_tumor_volume_median,
                                                      "Variance":mouse_tumor_volume_variance,"Standard Deviation":mouse_tumor_volume_standard_deviation,
                                                     "Standard Error":mouse_tumor_volume_standard_error})

mouse_tumor_volume_summary_statistics
```

```python
# Aggregation:
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume
# for each regimen
# Use pandas group function (.groupby) to group 'Drug Regimen' column and assign it a variable
drug_regimen_group = merged_mouse_dataset.groupby('Drug Regimen')
# Use pandas aggregation function (.agg) with the grouped variable to create a summary statistics table
aggregation_method = drug_regimen_group.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]
# Print to screen
aggregation_method.head(10)
```

## Mouse Drug Trial Chart Observation
It looks to me like this drug trial was not biased toward any one drug or any gender.  There are slight differences in the numbers, but for the most part, the drugs were tested at about the same rate and evenly amongst gender.  I would say that for this trial to be trusted, it would need to administer the exact same amount of test measurements per drug with an even split amongst gender.  For example, if there were 230 tests administered of Capomulin, then there should be 230 tests measurements for all the other drugs.  Further, if there were 230 measurements per drug regimen, then there should be 115 tests per male and 115 tests per female across all drugs.  I would think that this would provide the most accurate picture of what these drugs are capable of and which one is more effective at achieving its goal.

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
capomulin = merged_mouse_dataset.loc[merged_mouse_dataset["Drug Regimen"] == "Capomulin",:]
ramicane = merged_mouse_dataset.loc[merged_mouse_dataset["Drug Regimen"] == "Ramicane", :]
infubinol = merged_mouse_dataset.loc[merged_mouse_dataset["Drug Regimen"] == "Infubinol", :]
ceftamin = merged_mouse_dataset.loc[merged_mouse_dataset["Drug Regimen"] == "Ceftamin", :]
```

```python
# Capomulin:
# Start by getting the last (greatest) timepoint for each mouse
capomulin_last_greatest = capomulin.groupby('Mouse ID').max()['Timepoint']
capomulin_dataframe = pd.DataFrame(capomulin_last_greatest)
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
capomulin_merge = pd.merge(capomulin_dataframe, merged_mouse_dataset, on=("Mouse ID","Timepoint"),how="left")
capomulin_merge.head()
```

```python
# Quantitatively determine capomulin outliers
capomulin_tumor = capomulin_merge["Tumor Volume (mm3)"]

quartile = capomulin_tumor.quantile([.25,.5,.75])
lower_quartile = quartile[0.25]
upper_quartile = quartile[0.75]
iqr = upper_quartile-lower_quartile
lower_bound = lower_quartile - (1.5*iqr)
upper_bound = upper_quartile + (1.5*iqr)
print(f"Capomulin treatment has tumor volume outlier values that are below {lower_bound} and values above {upper_bound}.")
```

```python
# Ramicane:
ramicane_last_greatest = ramicane.groupby('Mouse ID').max()['Timepoint']
ramicane_dataframe = pd.DataFrame(ramicane_last_greatest)
ramicane_merge = pd.merge(ramicane_dataframe, merged_mouse_dataset, on=("Mouse ID","Timepoint"),how="left")
ramicane_merge.head()
```

```python
ramicane_tumor = ramicane_merge["Tumor Volume (mm3)"]

quartile = ramicane_tumor.quantile([.25,.5,.75])
lower_quartile = quartile[0.25]
upper_quartile = quartile[0.75]
iqr = upper_quartile-lower_quartile
lower_bound = lower_quartile - (1.5*iqr)
upper_bound = upper_quartile + (1.5*iqr)
print(f"Ramicane treatment has tumor volume outliers that are values below {lower_bound} and values above {upper_bound}.")
```

## Line and Scatter Plots

```python
# Line Plot:
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_line = capomulin.loc[capomulin["Mouse ID"] == "g288",:]
capomulin_line.head()
```

```python
# Define variable that will hold x axis data
x_axis = capomulin_line["Timepoint"]
# Define variable that will hold the y axis data
tumor_volume = capomulin_line["Tumor Volume (mm3)"]
# Create a title for the line chart
plt.title('Capomulin treatmeant: Volume over Time of Mouse g288')
# Use matplotlibe with the x and y variables established to plot the chart
plt.plot(x_axis, tumor_volume,linewidth=2, markersize=12)
# X axis label
plt.xlabel('Timepoint')
# Y axis label
plt.ylabel('Tumor Volume')
# Print to screen
plt.show()
```

```python
# Scatter Plot:
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Use pandas function groupby with 'Mouse ID' to find the mean in the capomulin drug treatment 
capomulin_average = capomulin.groupby(['Mouse ID']).mean()
# Use matplotlib to create a scatter plot of wieght to tumor volume
plt.scatter(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])
# Use matplotlib function .title and create a unique title for the chart
plt.title('Average Tumor Volume vs. Mouse Weight')
# Use matplotlib functions .xlabel and .ylabel to label the chart
plt.xlabel('Weight in grams')
plt.ylabel('Tumor Volume in Cubic Millimeter')
# Print to screen
plt.show()
```

## Correlation and Regression

```python
# Correlation Coefficient:
# Find the correlation coefficient/linear regression model for mouse weight vs average tumor volume in the Capomulin regimen
# Use scipy.stats function Pearson Correlation and define a variable to hold the results
capomulin_correlation=round(st.pearsonr(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])[0],2)
# Print a statement with the results
print(f"The correlation between mouse weight in grams and the average tumor volume in cubic millimeters is {capomulin_correlation}")
```

```python
# Linear Regression:
# Use scipy stats, linear regression, function (.linregress) and define a variable to hold the value
capomulin_linear=st.linregress(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])
capomulin_linear
```

```python
# Chart
# Define variables that hold the values of the linear Regression Results
capomulin_slope = 0.9544396890241045
capomulin_intercept = 21.552160532685015

y_values = capomulin_average['Weight (g)']*capomulin_slope+capomulin_intercept
# Create the scatter plot
plt.scatter(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])
# Create the line plot with matplotlib that will overlay the scatter plot
plt.plot(capomulin_average['Weight (g)'],y_values,color="red")
plt.xlabel('Weight in Grams')
plt.ylabel('Tumor Size in Cubic Millimeters')
# Print to screen
plt.show()
```
